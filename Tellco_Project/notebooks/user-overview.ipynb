```python
# Identify the top 3 manufacturers
top_3_manufacturers = df['Handset Manufacturer'].value_counts().head(3).index
print(top_3_manufacturers)
```

```python
# Dictionary to store the top 5 handsets for each manufacturer
top_handsets_per_manufacturer = {}

# Loop through each of the top 3 manufacturers
for manufacturer in top_3_manufacturers:
    # Filter the dataset for the current manufacturer
    filtered_df = df[df['Handset Manufacturer'] == manufacturer]
    
    # Find the top 5 handsets for the current manufacturer
    top_handsets = filtered_df['Handset Type'].value_counts().head(5)
    
    # Store the result in the dictionary
    top_handsets_per_manufacturer[manufacturer] = top_handsets

# Display the top handsets per manufacturer
for manufacturer, handsets in top_handsets_per_manufacturer.items():
    print(f"\nTop 5 Handsets for {manufacturer}:")
    print(handsets)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data for plotting
labels = list(top_handsets_per_manufacturer[top_3_manufacturers[0]].index)
x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Plot each manufacturer's top 5 handsets
for i, manufacturer in enumerate(top_3_manufacturers):
    ax.bar(x + i*width, top_handsets_per_manufacturer[manufacturer].values, width, label=manufacturer)

# Add some text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Handset Type')
ax.set_ylabel('Number of Users')
ax.set_title('Top 5 Handsets per Top 3 Manufacturers')
ax.set_xticks(x + width)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Select numerical features for PCA
numerical_features = ['Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)', 
                      'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 
                      'Other DL (Bytes)']

# Standardize the features
X = df[numerical_features].fillna(0)  # Fill missing values with 0 or another method
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()
```

```python
# Scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--', color='b', label='Explained Variance Ratio')
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='-', color='r', label='Cumulative Explained Variance')

plt.title('Scree Plot of PCA')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio / Cumulative Explained Variance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

# Apply K-Means clustering
def apply_kmeans(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(X)
    return clusters, kmeans

# Define the range for the number of clusters
cluster_range = range(1, 11)
sse = []  # Sum of squared errors

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)  # Sum of squared distances to closest cluster center
```

```python
# Elbow Method Plot
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, sse, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.grid(True)
plt.show()
```
