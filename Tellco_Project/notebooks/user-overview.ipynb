```python
# Aggregating data per user (MSISDN/Number)
aggregated_data = df.groupby('MSISDN/Number').agg(
    xdr_sessions=('Bearer Id', 'count'),  # Counting sessions per user
    session_duration=('Dur. (ms)', 'sum'),  # Summing the session duration
    total_download_data=('Total DL (Bytes)', 'sum'),  # Summing total download data
    total_upload_data=('Total UL (Bytes)', 'sum'),  # Summing total upload data
)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define the engagement metrics to normalize
engagement_metrics = ['xdr_sessions', 'session_duration', 'total_MB_volume']

# Normalize the engagement metrics using MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(aggregated_data[engagement_metrics])

# Convert normalized data into a DataFrame
normalized_df = pd.DataFrame(normalized_data, columns=engagement_metrics)

# Display the first few rows of normalized data
print(normalized_df.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Using the elbow method to find the optimal number of clusters
wcss = []  # Within-cluster sum of squares

# Test for cluster sizes from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(normalized_df)
    wcss.append(kmeans.inertia_)  # Inertia is WCSS

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K', fontsize=16)
plt.xlabel('Number of clusters (K)', fontsize=14)
plt.ylabel('WCSS', fontsize=14)
plt.show()
```

```python
# Let's assume the optimal number of clusters from the elbow method is 5
optimal_k = 5

# Apply K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(normalized_df)

# Add the cluster labels to the original data
aggregated_data['engagement_cluster'] = clusters

# Display the first few rows of data with clusters
print(aggregated_data.head())
```
